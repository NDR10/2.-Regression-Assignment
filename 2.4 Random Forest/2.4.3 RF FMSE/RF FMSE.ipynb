```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first = True)
dataset
```

```python
dataset.columns
```

```python
Input = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
Input
```

```python
Output = dataset[['charges']]
Output
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(Input, Output, test_size = 0.3, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
Regressor = RandomForestRegressor(n_estimators = 100, criterion = "friedman_mse")
Regressor = Regressor.fit(X_train, Y_train)
```

```python
Y_pred = Regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(Y_test, Y_pred)
r_score
```

```python
import pickle
filename = "Python_program_RF_FMSE.sav"
pickle.dump (Regressor, open(filename, "wb"))
```
